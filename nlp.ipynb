```python
from bs4 import BeautifulSoup
import shutil
import sys
```

```python
with open('export/messages.html', encoding='utf-8') as file:
    shutil.copyfileobj(file, sys.stdout)
```

```python
soup = BeautifulSoup('export/messages.html', "html.parser")
messages = soup.find_all("div", class_="text")
```

```python
import re
from html.parser import HTMLParser

def fix_errors(text):
    # Создаем словарь с ошибочными и правильными словами
    errors_dict = {
        "салем": "салем",
        "бүгін": "бугин",
        "спорт": "спорт",
        "Қазақстан" : "казахстан",
        "жаңа" : "жана"
    }
    for error, correct in errors_dict.items():
        text = text.replace(error, correct)

    return text
```

```python

with open("export/messages.html", "w") as parallel_file:
    for message in messages:
        text = message.text.strip()
        fixed_text = fix_errors(text)
        parallel_file.write(text + "\t" + fixed_text + "\n")

parallel_file.close()
```

```python

def parse_html(html_string):
    
    messages = []

    parser = HTMLParser()
    parser.feed(html_string)
    parser.close()


    for tag in parser.tag_data:
        if tag[0] == "div" and "class" in tag[1] and "message" in tag[1]["class"]:
            text = tag[2].strip()
            messages.append(text)
    
    return messages
```

```python

def create_parallel_text(html_string):
    messages = parse_html(html_string)

    with open("parallel.txt", "w") as parallel_file:
        for message in messages:
            text = message.text.strip()
            corrected_text = fix_errors(text)

        parallel_file.write(text + "\t" + corrected_text + "\n")

print(parallel_file)
```
